```python
import matplotlib.pyplot as plt

vehicle_types = ['SUV', 'Sedan', 'Truck', 'Van']
expenditure = [25000, 30000, 20000, 15000]  # Expenditure for each vehicle type during recession

colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)  # explode the 1st slice
labels = vehicle_types

plt.figure(figsize=(8, 6))
plt.pie(expenditure, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.title('Total Advertisement Expenditure by Vehicle Type During Recession Period')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
