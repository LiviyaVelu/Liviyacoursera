```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year': [2016, 2017, 2018, 2019, 2020, 2021],
    'GDP': [18.6, 19.2, 20.1, 20.8, 19.5, 21.0],
    'Recession_Period': ['No', 'No', 'No', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)

recession_data = df[df['Recession_Period'] == 'Yes']
non_recession_data = df[df['Recession_Period'] == 'No']

fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

axes[0].plot(recession_data['Year'], recession_data['GDP'], marker='o', color='r')
axes[0].set_title('GDP During Recession Period')
axes[0].set_ylabel('GDP')

axes[1].plot(non_recession_data['Year'], non_recession_data['GDP'], marker='o', color='g')
axes[1].set_title('GDP During Non-Recession Period')
axes[1].set_ylabel('GDP')
axes[1].set_xlabel('Year')

plt.tight_layout()
plt.show()
```
